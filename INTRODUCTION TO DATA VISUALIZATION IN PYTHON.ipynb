BASIC DATA VISUALIZATION

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# load the dataset

df = pd.read_csv(r"C:\Users\COISA COMPUTERS\OneDrive\Documents\PyTHON\student_marks.csv")
df.head(5)
```

```python
#1 use a bargraph to visualize perfomance per subject

plt.bar(df["Subject"], df["Marks"], color= 'red')
plt.title("Student marks per subject")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.show()


```

2 Line Chart

Line charts show trends over time (or ordered categories).
Letâ€™s pretend these are marks across tests:

```python
tests = ["Test1", "Test2", "Test3", "Test4", "Test5"]
marks = [65, 70, 80, 85, 90]



plt.plot(tests, marks, marker="o", color="yellow")
plt.title("Performance Over Tests")
plt.xlabel("Tests")
plt.ylabel("Marks")
plt.show()
```

3 Histogram:Histograms show the distribution of continuous data.
Letâ€™s simulate marks for 50 students in Math.

```python
np.random.seed(42)
math_marks = np.random.normal(70, 10, 50)  # mean=70, std=10
print(math_marks)

plt.hist(math_marks, bins=10, color="orange", edgecolor="black")
plt.title("Distribution of Math Marks")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()
```

4 Pie Chart

Pie charts show proportions

```python
plt.pie(df["Marks"], labels=df["Subject"], autopct="%1.1f%%", startangle=140)
plt.title("Percentage of Marks by Subject")
plt.show()
```

5 Donut Chart

```python
plt.pie(df["Marks"], labels=df["Subject"], autopct="%1.1f%%", startangle=140)
# Add circle for donut effect
centre_circle = plt.Circle((0,0),0.70,fc="white")
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title("Donut Chart of Marks by Subject")
plt.show()
```
